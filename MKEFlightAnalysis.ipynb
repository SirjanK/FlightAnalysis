# MKE Flight Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
%matplotlib inline
```

```python
# load data
path = './*.csv'

# Use glob to get all CSV file paths
all_files = glob.glob(path)

# Read and concatenate all CSV files into a single DataFrame
df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
```

```python
# plot the number of flights by day
def get_num_flights(day):
    return len(df[(df["ORIGIN_CITY_NAME"] == "Milwaukee, WI") & (df["DAY_OF_MONTH"] == day)])
```

```python
days = range(1, 32)
num_flights = [get_num_flights(day) for day in days]
plt.figure(figsize=(12, 6))
plt.plot(days, num_flights, 'b-', label='Number of Flights')

# Mark day 24 with a red dot
plt.plot(24, get_num_flights(24), 'ro', markersize=10, label='XMas Eve')

# Customize the plot
plt.title('Number of Flights by Day of Month MKE December 2018-2023', fontsize=16)
plt.xlabel('Day of Month', fontsize=12)
plt.ylabel('Number of Flights', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Set x-axis ticks to show all days
plt.xticks(days)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
def compute_delay_probability_with_cancellation(filtered_df, delay):
    num_delayed_or_cancelled = len(filtered_df[(filtered_df["ARR_DELAY"] >= delay) | (filtered_df["CANCELLED"] == True) | (filtered_df["DIVERTED"] == True)])
    return num_delayed_or_cancelled / len(filtered_df)
```

```python
def compute_delay_probability_without_cancellation(filtered_df, delay):
    not_cancelled = filtered_df[(filtered_df["CANCELLED"] == False) & (filtered_df["DIVERTED"] == False)]
    num_delayed = len(not_cancelled[(not_cancelled["ARR_DELAY"] >= delay)])
    return num_delayed / len(not_cancelled)
```

```python
filtered_df = df[df["ORIGIN_CITY_NAME"] == "Milwaukee, WI"]
xmas_eve_df = filtered_df[filtered_df["DAY_OF_MONTH"] == 24]
max_delay = filtered_df['ARR_DELAY'].max()

# Generate delay values from 0 to max_delay
max_delay = 250 # override
delays = np.arange(0, max_delay + 1)

# Compute probabilities for each delay
december_probabilities = [compute_delay_probability_without_cancellation(filtered_df, delay) for delay in delays]
xmas_eve_probabilities = [compute_delay_probability_without_cancellation(xmas_eve_df, delay) for delay in delays]

# Create the plot
plt.figure(figsize=(14, 7))

# Plot December probabilities in blue
plt.plot(delays, december_probabilities, 'b-', label='December')

# Plot Xmas Eve probabilities in red
plt.plot(delays, xmas_eve_probabilities, 'r-', label='Xmas Eve')

# Find the probabilities at 1 hour (60 minutes) delay
december_prob_1h = december_probabilities[60]
xmas_eve_prob_1h = xmas_eve_probabilities[60]

# Draw horizontal dotted lines at 1 hour mark
plt.axhline(y=december_prob_1h, color='b', linestyle=':', xmin=0)
plt.axhline(y=xmas_eve_prob_1h, color='r', linestyle=':', xmin=0)

# Add points at (60, prob) for both lines
plt.plot(60, december_prob_1h, 'bo', markersize=8)
plt.plot(60, xmas_eve_prob_1h, 'ro', markersize=8)

# Add text annotations above the points
plt.annotate(f'{december_prob_1h:.2%}', (60, december_prob_1h), xytext=(0, 10), 
             textcoords='offset points', ha='center', va='bottom', color='b')
plt.annotate(f'{xmas_eve_prob_1h:.2%}', (60, xmas_eve_prob_1h), xytext=(0, 10), 
             textcoords='offset points', ha='center', va='bottom', color='r')

# Customize the plot
plt.title('Delay Probability (without Cancellations) vs. Delay Duration MKE 2018-2023', fontsize=16)
plt.xlabel('Delay (hours:minutes)', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Set y-axis to display probabilities as percentages
plt.ylim(0, 0.5)
plt.yticks(np.arange(0, 0.51, 0.1))
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))

# Add legend
plt.legend(fontsize=10)

# Adjust x-axis to show hours and minutes in 30-minute increments
def format_time(x, pos):
    hours = int(x // 60)
    minutes = int(x % 60)
    return f'{hours:02d}:{minutes:02d}'

plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(format_time))
plt.xticks(np.arange(0, max_delay + 1, 30), rotation=45)

# Add some padding to x-axis for better visibility
plt.xlim(0, max_delay * 1.05)

# Show the plot
plt.tight_layout()
plt.show()
```
